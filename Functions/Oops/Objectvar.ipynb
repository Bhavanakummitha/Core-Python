```python
class Train:
    def __init__(self, name, number, type_of_train, capacity, speed, origin, destination, departure_time):
        self.name = name                  
        self.number = number              
        self.type_of_train = type_of_train  
        self.capacity = capacity          
        self.speed = speed                
        self.origin = origin             
        self.destination = destination    
        self.departure_time = departure_time  

    def display(self):
        print("Train Name: {self.name")
        print("Train Number: {self.number")
        print("Type of Train: {self.type_of_train")
        print("Capacity: {self.capacity")
        print("Speed: {self.speed} km/h")
        print("Origin: {self.origin")
        print("Destination: {self.destination")
        print("Departure Time: {self.departure_time")
        print("-" * 30)

# Creating three Train objects
train1 = Train("Express A", "101", "Express", 300, 120, "City A", "City B", "10:00 AM")
train2 = Train("Local B", "202", "Local", 150, 60, "City C", "City D", "11:00 AM")
train3 = Train("Superfast C", "303", "Superfast", 500, 150, "City E", "City F", "12:00 PM")

# Displaying the details of each train
train1.display()
train2.display()
train3.display()


```

```python
class Student:
    def __init__(self, name, student_id, age, major, gpa):
        self.name = name              
        self.student_id = student_id     
        self.age = age                  
        self.major = major            
        self.gpa = gpa                   

    def display(self):
        print("Student Name: self.name")
        print("Student ID: self.student_id")
        print("Age: self.age")
        print("Major: self.major")
        print("GPA: self.gpa")
        print("-" * 30)

# Creating three Student objects
student1 = Student("Alice Smith", "S001", 20, "Computer Science", 3.8)
student2 = Student("Bob Johnson", "S002", 21, "Mathematics", 3.5)
student3 = Student("Charlie Brown", "S003", 22, "Physics", 3.9)

# Displaying the details of each student
student1.display()
student2.display()
student3.display()
```

```python
class College:
    def __init__(self, name, location, established_year, num_courses, num_students, college_type, accreditation, average_fee):
        self.name = name                      
        self.location = location              
        self.established_year = established_year  
        self.num_courses = num_courses        
        self.num_students = num_students      
        self.college_type = college_type      
        self.accreditation = accreditation    
        self.average_fee = average_fee       

    def display(self):
        print("College Name: self.name")
        print("Location: self.location")
        print("Established Year: self.established_year")
        print("Number of Courses: self.num_courses")
        print("Number of Students: self.num_students")
        print("Type of College: self.college_type")
        print("Accreditation: self.accreditation")
        print("Average Fee:self.average_fee")
        print("-" * 30)

# Creating three College objects
college1 = College("ABC College", "New York", 1995, 10, 1200, "Private", "Accredited", 15000)
college2 = College("XYZ University", "Los Angeles", 2000, 20, 3000, "Public", "Accredited", 10000)
college3 = College("LMN Institute", "Chicago", 1985, 15, 800, "Private", "Not Accredited", 12000)

# Displaying the details of each college
college1.display()
college2.display()
college3.display()
```
